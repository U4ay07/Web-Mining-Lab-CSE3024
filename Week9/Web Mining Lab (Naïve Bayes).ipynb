```python
columns = ['train', 'class']
rows = []
rows = [['president-nod-for-lokpal-bill', 'India'], 
 ['india-scraps-vvip-chopper-deal-with-agusta-westland', 'India'],
 ['mdmk-to-be-part-of-bjp-led-nda', 'India'],
 ['ex-envoy-hardeep-puri-joins-bjp', 'India'],
 ['modi-to-address-rally-in-panaji', 'India'],
 ['church-not-against-modi-bishop', 'India'],
 ['aap-government-wins-confidence-vote', 'India'],
 ['aap-retains-jhaadu-as-its-symbol', 'India'],
 ['modi-accepts-ramdevs-terms-for-support', 'India'],
 ['violence-mars-poll-in-bangladesh', 'Bangladesh'],
 ['sheikh-hasina-set-to-form-govt-again', 'Bangladesh'],
 ['four-killed-in-postpoll-violence-in-bangladesh', 'Bangladesh'],
 ['maldives-president-coming-today', 'Others'],
 ['bhutan-king-arrives-on-5-day-visit', 'Others'],
 ['seemandhra-bandh-hits-ap-tn', 'India']]
```

```python
training_data = pd.DataFrame(rows, columns=columns)
training_data
```

```python
p_india = 10 / 15
p_bangladesh = 3 / 15
p_others = 2 / 15
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

### For India Class

```python
India_docs = [row['train'] for index,row in training_data.iterrows() if row['class'] == 'India']
vec_i = CountVectorizer()
X_i = vec_i.fit_transform(India_docs)
freq_wi = pd.DataFrame(X_i.toarray(), columns=vec_i.get_feature_names())
freq_wi
```

### For Bangladesh Class

```python
Bangladesh_docs = [row['train'] for index,row in training_data.iterrows() if row['class'] == 'Bangladesh']
vec_b = CountVectorizer()
X_b = vec_b.fit_transform(Bangladesh_docs)
freq_wb = pd.DataFrame(X_b.toarray(), columns=vec_b.get_feature_names())
freq_wb
```

### For Other Class

```python
Others_docs = [row['train'] for index,row in training_data.iterrows() if row['class'] == 'Others']
vec_o = CountVectorizer()
X_o = vec_o.fit_transform(Others_docs)
freq_wo = pd.DataFrame(X_o.toarray(), columns=vec_o.get_feature_names())
freq_wo
```

## Frequency of words in: 

### India class

```python
word_list_i = vec_i.get_feature_names(); 
count_list_i = X_i.toarray().sum(axis=0) 
freq_i = dict(zip(word_list_i,count_list_i))
print(freq_i)
```

### Bangladesh Class

```python
word_list_b = vec_b.get_feature_names(); 
count_list_b = X_b.toarray().sum(axis=0) 
freq_b = dict(zip(word_list_b,count_list_b))
print(freq_b)
```

### Other Class

```python
word_list_o= vec_o.get_feature_names(); 
count_list_o = X_o.toarray().sum(axis=0) 
freq_o = dict(zip(word_list_o,count_list_o))
print(freq_o)
```

```python
docs = [row['train'] for index,row in training_data.iterrows()]
vec = CountVectorizer()
X = vec.fit_transform(docs)
total_features = len(vec.get_feature_names())
total_features
```

```python
#To find, ‘total count of all features in a class ’,
total_cnts_features_i = count_list_i.sum(axis=0)
total_cnts_features_b = count_list_b.sum(axis=0)
total_cnts_features_o = count_list_o.sum(axis=0)
```

## Probablity of words according to Laplacian correction model in: 

### India class

```python
prob_wi_i=[]
for word,count in zip(word_list_i,count_list_i):
    prob_wi_i.append((count + 1) / (total_cnts_features_i + total_features))
print(dict(zip(word_list_i,prob_wi_i)))
```

### Bangladesh Class

```python
prob_wb_b=[]
for word,count in zip(word_list_b,count_list_b):
    prob_wb_b.append((count + 1) / (total_cnts_features_b + total_features))
print(dict(zip(word_list_b,prob_wb_b)))
```

### Other Class

```python
prob_wo_o=[]
for word,count in zip(word_list_o,count_list_o):
    prob_wo_o.append((count + 1) / (total_cnts_features_i + total_features))
print(dict(zip(word_list_o,prob_wo_o)))
```
